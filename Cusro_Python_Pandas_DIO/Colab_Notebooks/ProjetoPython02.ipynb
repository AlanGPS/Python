# Tuplas

```python
# As tuplas usam parênteses como sintaxe.
tp = ("banana", "maça", 10, 50)
tp
```

```python
tp[0]
```

```python
# Diferente de LISTA, a TUPLA é imutável.
tp[2] = "laranja"
```

```python
tp.count("maça")
```

```python
tp[0:2]
```

# Dicionários

```python
# Para criar um DICIONÁRIO, utilizamos chaves {}.
# O dicionário trabalha com o conceito: Chave e Valor
dc = {"maça":20,"banana":10,"laranja":15,"uva":5}
dc
```

```python
dc["maça"]
```

```python
dc["maça"] = 25
dc
```

```python
dc.keys()
```

```python
dc.values()
```

```python
# Verifica no dicionário e se não existir, ele adiciona.
dc.setdefault("limão",22)
```
